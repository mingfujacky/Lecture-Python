### Series city_revenues

```python
import pandas as pd
```

```python
city_revenues = pd.Series(
    [4200, 8000, 6500],
    index = ['Amsterdam', 'Toronto', 'Tokyo']   # create series by list
)
city_revenues
```

```python
print(city_revenues['Toronto'])
print(city_revenues[1])  # Toronto's index is 1
print(city_revenues[0])
print(city_revenues[-1])

print(city_revenues.sum())
# city_revenues.min()
# city_revenues.max()
# city_revenues.mean()
```

### Series city_employee_count

```python
city_employee_count_data = {'Amsterdam': 5, 'Tokyo': 8}   # create series by dict
city_employee_count = pd.Series(city_employee_count_data)
city_employee_count
```

### DataFrame city_data

```python
city_data = pd.DataFrame({
    'revenue': city_revenues,
    'employee_count': city_employee_count
})
city_data
```

### Merge DataFrame

```python
further_city_data = pd.DataFrame(
    {
        'revenue': [7000, 3400],
        'employee_count': [2, 2]
    },
    index=['New York', 'Barcelona']
)
further_city_data
```

```python
all_city_data = pd.concat([city_data, further_city_data], sort=False)
all_city_data
```

### Join DataFrame

```python
city_countries = pd.DataFrame({
    'country': ['Holland', 'Japan', 'Holland', 'Canada', 'Spain'],
    'capital': [1, 1, 0, 0, 0]
}, index=['Amsterdam', 'Tokyo', 'Rotterdam', 'Toronto', 'Barcelona'])
city_countries
```

```python
cities = pd.concat([all_city_data, city_countries], axis=1, sort=False, join='inner')
cities
```

```python
countries = pd.DataFrame({
    'population_millions': [17, 127, 37],
    'continent': ['Europe', 'Asia', 'North America']
}, index=['Holland', 'Japan', 'Canada'])
countries
```

```python
pd.merge(cities, countries, left_on='country', right_index=True)
```

```python
pd.merge(cities, countries, left_on='country', right_index=True, how='left')
```
